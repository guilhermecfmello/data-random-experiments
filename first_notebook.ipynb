```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')

# Check the first few rows
print(iris.head())
```

```python
print("Mean:", iris['sepal_length'].mean())
print("Median:", iris['sepal_length'].median())
print("Mode:", iris['sepal_length'].mode()[0])
```

```python
print("Standard Deviation:", iris['petal_length'].std())
print("Variance:", iris['petal_length'].var())
```

```python
sepal_width_range = iris['sepal_width'].max() - iris['sepal_width'].min()
print("Range of Sepal Width:", sepal_width_range)
```

```python
Q1 = iris['petal_width'].quantile(0.25)
Q3 = iris['petal_width'].quantile(0.75)
IQR = Q3 - Q1
print("IQR for Petal Width:", IQR)
```

```python
plt.figure(figsize=(8, 6))
plt.hist(iris['sepal_length'], bins=30, edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=iris['petal_length'])
plt.title('Boxplot of Petal Length')
plt.show()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
# Select only the numerical columns for correlation
numerical_columns = iris.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr = numerical_columns.corr()

# Plot the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
